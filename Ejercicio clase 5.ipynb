#Matplotlib: Es una biblioteca de Python utilizada para crear gráficos y visualizaciones de datos de alta calidad. Es una herramienta poderosa para convertir datos en gráficos visualmente atractivos y comprensibles.

Proporciona una interfaz de alto nivel para crear gráficos estadísticos atractivos y fáciles de interpretar.
```python
import matplotlib.pyplot as plt
import seaborn as sns

#%matplotlib inline puede usar si  salta error se usa para observa graficos
```

```python
plt.figure(figsize=(10,5))#tamaño del grafico
```

```python
#1er grafico
plt.figure(1,figsize=(10,5))
sns.countplot(y='Gender',data=df) #countplot para grafico de barras
plt.show()
```

```python
sns.displot(df['Spending Score (1-100)'])  #distribucion de variable  displot
# como una campana gaussina
```

```python
sns.distplot(df['Spending Score (1-100)'], label='Spending Score',color='b')# sns.displot distribuye la variable en una grafica
sns.distplot(df['Age'],label='Age',color='r')
sns.distplot(df['Annual Income (k$)'],label='Annual Income (k$',color='g')

plt.legend(loc='upper right')
plt.title('analisis de datos')

plt.xlabel('Valores')
plt.ylabel('Densidad')

plt.show()
```

```python
sns.displot(df['Spending Score (1-100)'], kind='hist', kde=True, color='skyblue', edgecolor='black', bins=20)
plt.title('Distribución del Spending Score')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
#relacion
#concatenar dos columnas  usae concat
#scatter grafica de puntos
var='Annual Income (k$)'
data = pd.concat([df['Spending Score (1-100)'],df[var]],axis = 1)
data.plot.scatter(x=var,y='Spending Score (1-100)',ylim=(0,100),color='purple')
#hay personas que ganan menos gastan mas y personas que ganan mucho y gastan poco
```
