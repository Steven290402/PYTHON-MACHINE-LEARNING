```python
#ejemplo
estudiantes = {}

def mostrar_estudiantes():
    if estudiantes:
        print("\nLista de estudiantes:")
        for nombre, materias in estudiantes.items():
            print(f"Estudiante: {nombre}")
            for materia, calificacion in materias.items():
                print(f"  {materia}: {calificacion}")
    else:
        print("\nNo hay estudiantes registrados.")

activar = True

while activar:
    print("\nMenú:")
    print("1. Añadir estudiante y calificaciones")
    print("2. Mostrar estudiantes y calificaciones")
    print("3. Calcular promedio de un estudiante")
    print("4. Salir")

    opcion = input("Elige una opción: ")

    if opcion == '1':
        nombre = input("Nombre del estudiante: ")
        estudiantes[nombre] = {}
        while True:
            materia = input("Introduce el nombre de la materia (o escribe 'fin' para terminar): ")
            if materia.lower() == 'fin':
                break
            calificacion = float(input(f"Introduce la calificación para {materia}: "))
            estudiantes[nombre][materia] = calificacion
        print(f"Las calificaciones de {nombre} han sido registradas.")
    elif opcion == '2':
        mostrar_estudiantes()
    elif opcion == '3':
        nombre = input("Introduce el nombre del estudiante para calcular el promedio: ")
        if nombre in estudiantes:
            materias = estudiantes[nombre]
            if materias:
                promedio = sum(materias.values()) / len(materias)
                print(f"El promedio de {nombre} es: {promedio:.2f}")
            else:
                print(f"{nombre} no tiene calificaciones registradas.")
        else:
            print(f"No se encontró al estudiante {nombre}.")
    elif opcion == '4':
        activar = False
        print("¡Hasta luego!")
    else:
        print("Opción no válida, por favor elige una opción del menú.")

print("Programa finalizado.")
```

```python
#los operadores * y ** en los argumentos de una función permiten manejar un número variable de argumentos.
#Son útiles para hacer funciones más flexibles
#El operador * permite a una función aceptar un número arbitrario de argumentos posicionales, que serán
#recogidos en forma de una tupla. Esto es útil cuando no sabes cuántos argumentos vas a recibir

def promedio_notas(nombre,*notas):
  print(f"El estudiante {nombre}")
  if notas:
    promedio =max(notas)
    print(f"la nota mas alta es  {promedio}")
  else:
    print("no tiene notas")
```

```python
class Carro:
    def __init__(self, marca, modelo, color, velocidad, cilindraje, t_combustible):
        self.marca = marca
        self.modelo = modelo
        self.color = color
        self.velocidad = velocidad
        self.cilindraje = cilindraje
        self.t_combustible = t_combustible
        self.odometro = 0  # Valor inicial del odómetro

    def mostrar(self):
        print(f"Marca: {self.marca}")
        print(f"Modelo: {self.modelo}")
        print(f"Color: {self.color}")
        print(f"Velocidad: {self.velocidad} km/h")
        print(f"Cilindraje: {self.cilindraje} L")
        print(f"Tipo de combustible: {self.t_combustible}")

    def ubicacion(self, ubicacion):
        self.ubicacion = ubicacion
        print(f"Ubicación: {self.ubicacion}")

    def mostrar_odometro(self):
        print(f"Odómetro: {self.odometro} km")

    def actualizar_odometro(self, kilometros):
        """Actualiza el odómetro con un valor dado."""
        if kilometros >= self.odometro:  #no retroceder el odómetro
            self.odometro = kilometros
        else:
            print("el valor del odómetro es no reducible.")


    def incrementar_odometro(self, kilometros):
         """Incrementa el valor del odómetro."""
         if kilometros > 0:
            self.odometro += kilometros
         else:
            print("El incremento debe ser positivo.")

```

```python
carro_1 = Carro("chevrolet", "sail", "Rojo", 300, 1.7, "Gasolina")
carro_1.mostrar()
carro_1.ubicacion("Zona centro")
carro_1.mostrar_odometro()

# Actualización del odómetro
carro_1.actualizar_odometro(20)
carro_1.mostrar_odometro()
print("\n")
carro_2 = Carro("jeep", "4x4", "negro", 200, 1.7, "Gasolina")
carro_2.mostrar()
carro_2.ubicacion("Zona centro")
carro_2.mostrar_odometro()

carro_1.actualizar_odometro(50)
carro_1.mostrar_odometro()
```
